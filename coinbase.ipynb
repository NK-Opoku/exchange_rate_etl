```python
url = 'https://api.coinbase.com/v2/exchange-rates'
response = requests.get(url)
ex_rates = json.loads(response.text)
ex_rates 
```

```python
# Getting rate for specicific currencies
url = 'https://api.coinbase.com/v2/exchange-rates?currency=BTC'
response = requests.get(url)
ex_rates = json.loads(response.text)
ex_rates 
ex_rates['data']['rates']['USD']
```

```python
#  querying historical data pairing two different currencies
url = 'https://api.coinbase.com/v2/prices/USD-GHS/spot?date=2022-01-31'
response = requests.get(url)
spot_price = json.loads(response.text)
spot_price 
spot_price['data']['rates']['USD']
```
